#### Chart - 5

````python
# Chart - 5 visualization code
# Create a pivot table for arrival month vs. hotel type
pivot_table = a1.pivot_table(index='arrival_date_month', columns='Hotels', values='is_canceled', aggfunc='count')

# Create a heatmap
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt='g')

# Set the plot title and labels
plt.title("Arrival Month vs. Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Arrival Month")

# Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

we create a pivot table to analyze the relationship between arrival month, hotel type, and the number of bookings.

##### 2. What is/are the insight(s) found from the chart?

The heatmap() function from seaborn is used to create a heatmap visualization of the pivot table, where darker colors represent higher booking counts.
